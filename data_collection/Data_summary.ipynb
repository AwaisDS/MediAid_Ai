```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mediaid_dataset.csv')
```

```python
df.head()
```

```python
df['disease'].value_counts()
```

```python
print(df.dtypes)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
        
# Plot 1: Disease Frequency
plt.figure()
df['disease'].value_counts().plot(kind='bar')
plt.title("Disease Frequency")
plt.xlabel("Disease")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Region Distribution
plt.figure()
df['region'].value_counts().plot(kind='bar')
plt.title("Region Distribution")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python

# Plot 3: Age Distribution Histogram
plt.figure()
df['age'].plot(kind='hist', bins=20)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
